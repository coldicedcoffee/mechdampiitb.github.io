```python
def extract_first_letters(course_code):
    # Using regular expression to extract the first letters
    match = re.match(r'[a-zA-Z]+', course_code)
    if match:
        first_letters = match.group()
        return first_letters
    else:
        return None  # Return None if no letters are found
```

```python
def generate_file_name(course_info):
    course_name = course_info['Name of the Course:'].split(' ')[0].lower()
    
    prof_name = course_info['Name of the Instructor:\n\nFormat : Prof. <firstname> <lastname>'].replace('.', '').replace(' ', '_').lower()
    
    date_today = str(date.today()).replace(' ', '') + '-'
    
    file_name = date_today + course_name + '_' + prof_name + '.md'
#     print(file_name)
    return file_name

def create_file(file_name, path = "../_posts/Course Reviews/New_reviews"):
    file_path = os.path.join(path, file_name)
#     print(file_path)
    file = open(file_path, "w")
    return file_path

def write_header(file_path, course_info):
    dept = get_dept(course_info['Name of the Course:'])
    print(dept)
    with open(file_path, 'a') as file:
        file.write('---\n')
        file.write('layout: post\n')
        file.write('title: ' + "\"" + course_info['Name of the Course:'] + "\"\n")
        
        categories = course_info.get('Please select some of the following tags that would be relevant to the particular course', '')
        if pd.notna(categories):
            file.write(categories + '\n\n')
        
        file.write('tags: [courses]\n')
        file.write('image: '+ image_dict[dept] +  '\n')
        file.write('featured: false\n')
        file.write('hidden: false\n')
        file.write('---\n\n')
```

```python
import pandas as pd

def get_dept(course_name):
    course_dept = extract_first_letters(course_name.split(' ')[0])
    return course_dept

def write_content(file_path, course_info):
    with open(file_path, 'a') as file:
        file.write("#### Instructor\n")
        instructor_name = course_info.get('Name of the Instructor:\n\nFormat : Prof. <firstname> <lastname>', '')
        if pd.notna(instructor_name):
            file.write(instructor_name + '\n\n')
        
        section = course_info.get('Section:')
        if pd.notna(section):
            file.write("#### Section\n")
            file.write(section + '\n\n')
        
        semester_year = course_info.get('Semester/Year', '')
        if pd.notna(semester_year):
            file.write('#### Semester\n')
            file.write(semester_year + '\n\n')
        
        course_difficulty = course_info.get('Course Difficulty:', '')
        if pd.notna(course_difficulty):
            file.write('#### Course Difficulty\n')
            file.write(course_difficulty + '\n\n')
        
        time_commitment = course_info.get('Time commitment needed:', '')
        if pd.notna(time_commitment):
            file.write('#### Time commitment needed\n')
            file.write(time_commitment + '\n\n')
        
        grading_statistics = course_info.get('Grading Statistics:', '')
        if pd.notna(grading_statistics):
            file.write('#### Grading Statistics\n')
            file.write(grading_statistics + '\n\n')
        
        attendance_policy = course_info.get('Attendance Policy:')
        if pd.notna(attendance_policy):
            file.write("#### Attendance Policy\n")
            file.write(attendance_policy + '\n\n')
        
        teaching_style = course_info.get('Teaching Style:', '')
        if pd.notna(teaching_style):
            file.write('#### Teaching Style\n')
            file.write(teaching_style + '\n\n')
        
        feedback_assignments = course_info.get('Feedback on Assignments/ Tutorials/ Projects', '')
        if pd.notna(feedback_assignments):
            file.write('#### Feedback on Assignments/ Tutorials/ Projects\n')
            file.write(feedback_assignments + '\n\n')
        
        feedback_exams = course_info.get('Feedback on Exams (Written Evaluation):', '')
        if pd.notna(feedback_exams):
            file.write('#### Feedback on Exams (Written Evaluation)\n')
            file.write(feedback_exams + '\n\n')

        future_tracks = course_info.get('Future Tracks:', '')
        if pd.notna(future_tracks):
            file.write('#### Future Tracks\n')
            file.write(future_tracks + '\n\n')
        
        course_importance = course_info.get('Course Importance:', '')
        if pd.notna(course_importance):
            file.write('#### Course Importance\n')
            file.write(course_importance + '\n\n')
        
        additional_details = course_info.get('Additional Details:', '')
        if pd.notna(additional_details):
            file.write('#### Additional Details\n')
            file.write(additional_details + '\n\n')
        
        contact_details = course_info.get('Please mention your Phone Number (if you want it to be displayed on the blog)', '')
        if pd.notna(contact_details):
            file.write('#### Contact Details\n')
            file.write(contact_details + '\n\n')
```

```python
def generate_readme(course_info):
    file_name = generate_file_name(course_info)
    
    file_path = create_file(file_name)
    write_header(file_path, course_info)
    write_content(file_path, course_info)
```

```python
for index, row in reviews.iterrows():
    generate_readme(row)
```
