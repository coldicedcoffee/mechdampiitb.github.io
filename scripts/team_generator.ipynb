```python
import pandas as pd
import numpy as np
```

```python
team_data = pd.read_csv("../assets/team_24_25.csv")

team_dict = team_data.T.to_dict()
```

```python
print(team_dict)
```

```python
import os

folder_path = "../assets/team_images_24"  # Specify the path to your folder

# Get a list of all files in the folder
file_list = os.listdir(folder_path)
# print(file_list)

extensions = ['.jpg', '.jpeg', '.png','.PNG','.JPG']
for key, values in team_dict.items():
    for extension in extensions:
        image_name = values["Name"] + extension
        if(os.path.exists(os.path.join(folder_path, image_name))):
            values['src'] = os.path.join(folder_path, image_name)
            values['src'] = values['src'].replace('\\', '/')
            print(values['src'])
```

```python
import jinja2

subs = jinja2.Environment( 
              loader=jinja2.FileSystemLoader('./')      
              ).get_template('profile.html').render(team=team_dict)
# lets write the substitution to a file
with open('test1.html','w', encoding="utf-8") as f: f.write(subs)
```
